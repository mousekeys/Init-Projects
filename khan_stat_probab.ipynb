```python
def get_column_length(df): #GET THE NUMBER OF COLUMNS
    return len(df.columns)

def get_count_unique(df): #GET THE NUMBER OF UNIQUE VALUES
    return df.nunique()

def get_col_type(df): #GET THE DATA TYPE OF EACH COLUMN OR SINGLE COLUMN
    return df.dtypes

def get_quantitative_qualitative(df): #COUNT OUT THE TYPE FO DATA IN DF
    return df.dtypes.value_counts()

def get_stats(df,info=''): #GET THE STATISTICS OF THE DATAFRAME
    if bool(info):
        print(info)
        return df.describe().loc[info]
    return df.describe()

```
